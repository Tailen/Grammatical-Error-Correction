### Inspect Token Length Distribution

```python
from transformers import T5Tokenizer
tokenizer = T5Tokenizer.from_pretrained('t5-base')
val_df[2] = val_df[0].apply(lambda x: len(tokenizer(x).input_ids))
val_df[3] = val_df[1].apply(lambda x: len(tokenizer(x).input_ids))
```

```python
# Check the distribution of input token lengths
val_df[2].describe()
```

```python
# Check the distribution of output token lengths
val_df[3].describe()
```

```python
print("95th percentile of input token length: {}".format(val_df[2].quantile(0.95)))
print("95th percentile of output token length: {}".format(val_df[3].quantile(0.95)))
print("99th percentile of input token length: {}".format(val_df[2].quantile(0.99)))
print("99th percentile of output token length: {}".format(val_df[3].quantile(0.99)))
```
